```python
from dotenv import load_dotenv
from langchain_openai.chat_models import ChatOpenAI
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
```

```python
load_dotenv()
model = ChatGoogleGenerativeAI(model = 'gemini-2.5-flash',temperature=0.0)
# intialize the OpenAI chat model
# model = ChatOepnAI(model='gpt-5-nano')
```

```python
#define Prompt template

template =  PromptTemplate.from_template(
    """ You are a helpfull assistant that answers Questions based on the provided context
    context: {context}
    Question : {Question}
    Answer: """
)
```

```python
# generate the system prompt
user_input={
    'context': 'The capital of India is New Delhi',
    'Question': 'What is the Capital of France'
}
```

```python
chain=template | model # create a chain that combines the template and the model
```

```python
#generate response using the chain
response=chain.invoke(user_input)

# user-> template -> model -> response    # step-by-step flow
```

```python
print(response)
print('---------------------')
print(response.content)
```
